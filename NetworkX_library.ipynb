## Δημιουργία επιπλέον χαρακτηριστικών για τους κόμβους

```python
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
```

```python
print(G.nodes[0])
```

### Ταξινόμηση κόμβων ανα βαθμό

```python
from operator import itemgetter
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)
print("Nodes by degree:")
for d in sorted_degree:
    print(d)
```

## Εύρεση betweenness και eigenvector centrality

```python
betweenness_dict = nx.betweenness_centrality(G)
eigenvector_dict = nx.eigenvector_centrality(G) 

nx.set_node_attributes(G, betweenness_dict, 'betweenness')
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')
```

```python
sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)
print("Nodes by betweenness centrality:")
for b in sorted_betweenness:
    print(b)
```

```python
sorted_eigenvector = sorted(eigenvector_dict.items(), key=itemgetter(1), reverse=True)
print("Nodes by eigenvector centrality:")
for e in sorted_eigenvector:
    print(e)
```

```python
for betweenness in sorted_betweenness: 
    degree = degree_dict[betweenness[0]]
    for e in sorted_eigenvector:
      if e[0] == betweenness[0]:
        eigen = e[1]
        break 
    print("Name:", betweenness[0], "| Betweenness Centrality:", betweenness[1], "| Eigenvector Centality", eigen,  "| Degree:", degree)
```

## Girvan-Newman community detection

```python
# Girvan Newman algorithm for community detection
number_of_communities = 2
num_of_cc = nx.number_connected_components(G)
while num_of_cc < number_of_communities:
    B = nx.edge_betweenness_centrality(G)
    edge = max(B.items(), key=itemgetter(1))[0]
    G.remove_edge(edge[0], edge[1])
    print(f'Removed edge: {edge}')
    num_of_cc = nx.number_connected_components(G)

# Colors for the nodes of each community
palette = ['yellow', 'cyan', 'lightgreen', 'lightgrey']
node_color = [None] * len(G)
cc = nx.connected_components(G)
for i,c in enumerate(cc):
    for node in c:
        node_color[node] = palette[i]

# Convert cc to list of communities
cc = nx.connected_components(G)
partition = list()
for c in cc:
    partition.append(list(c))


# Draw the graph using the colors 
plt.title('Girvan-Newman clustering')
nx.draw(G, with_labels=True, node_color=node_color)
plt.show()

# Metrics for the clustering
import networkx.algorithms.community as nx_comm

modularity = nx_comm.modularity(G, partition)
print(f'Modularity (Girvan-Newman): {modularity}')

coverage, performance = nx_comm.partition_quality(G, partition)  #coverage(G, partition)
print(f'Coverage (Girvan-Newman): {coverage}')
print(f'Performance (Girvan-Newman): {performance}')
```
